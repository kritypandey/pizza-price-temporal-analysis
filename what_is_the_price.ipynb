#DATA CLEANING

```python
#CHANGE DATATYPE
transaction['order_date']=pd.to_datetime(transaction['order_date'])
print(transaction.info())
```

```python
price_history['effective_date']=pd.to_datetime(price_history['effective_date'])
print(price_history.info())
```

##Sort before merging


```python
#sort transaction by order_date
transaction.sort_values(by='order_date')
```

```python
price_history.sort_values(by='effective_date')
```

##Apply ASOF Merge

```python
result= pd.merge_asof(transaction,price_history,left_on='order_date', right_on='effective_date', by='pizza_id', direction='backward')
```

```python
result.head()
```

###FIND REVENUE

```python
import numpy as np
```

```python
#REVENUE
result["revenue"] = result["price"] * result["quantity"]
```

```python
result.head()
```

##Round Down to Nearest Integer
####TOTAL REVENUE

```python
#total revenue


total_revenue = np.floor(result['revenue'].sum())

print(total_revenue)
```
